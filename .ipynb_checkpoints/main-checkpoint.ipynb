```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline 
def pca(X,k):#k is the components you want
    #mean of each feature
    n_samples, n_features = X.shape
    mean=np.array([np.mean(X[:,i]) for i in range(n_features)])
    #normalization
    norm_X=X-mean
    #scatter matrix
    scatter_matrix=np.dot(np.transpose(norm_X),norm_X)
    #Calculate the eigenvectors and eigenvalues
    eig_val, eig_vec = np.linalg.eig(scatter_matrix)
    eig_pairs = [(np.abs(eig_val[i]), eig_vec[:,i]) for i in range(n_features)]
    # sort eig_vec based on eig_val from highest to lowest
    eig_pairs.sort(reverse=True)
    # select the top k eig_vec
    feature=np.array([ele[1] for ele in eig_pairs[:k]])
    #get new data
    data=np.dot(X,np.transpose(feature))
    return data, feature

def image_to_vector(image):
    face_source_vector = np.reshape(image, (19 * 19,))
    return face_source_vector


```

```python

if __name__ == "__main__":
    all_train_face_path = os.listdir("face/train/")
    all_face_matrix = []
    for train_face_path in all_train_face_path:
        face_source = cv2.imread("face/train/" + train_face_path, 0)
        face_vector = image_to_vector(face_source)
        all_face_matrix.append(face_vector)
#     plt.subplot(121);plt.imshow(all_face_matrix[0])  # expects distorted color
#     plt.subplot(122);plt.imshow(all_face_matrix[0])  # expects true color
#     plt.show()
    all_distance = []
    max_k = 381
    for k in xrange(max_k):
        all_face_pca_matrix, pca_feature = pca(np.array(all_face_matrix), k + 1)
        pca_reconstruct_matrix = np.dot(all_face_pca_matrix, pca_feature)
        distance_L2 = np.mean(np.square(pca_reconstruct_matrix - all_face_matrix))
        all_distance.append(distance_L2)
    plt.figure(1) # 创建图表1
    plt.plot(range(max_k), all_distance)
    plt.show()
    print all_distance

        
   
```
